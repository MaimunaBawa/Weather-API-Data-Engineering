```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
```

```python
br_mg = pd.read_csv("data/csv/br_final.csv")
br_mg["year"] = pd.to_datetime(br_mg["year"], format="%Y")

#Filter for Minas Gerias


plt.figure(figsize= (12,8))
sb.lineplot(data= br_mg, x = "year", y = "million_60kgs_bag",hue="subdivision", marker = "o")
plt.title("Harvest Outcomes Across Regions in Brazil")
plt.xlabel("Year")
plt.ylabel("Million 60kgs_bag")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

How have harvest outcomes changed for Minas Gerais throughout the years?  

Throughout the year harvest outcome for Minas Gerais have been increasing with variations form year to year.There is a general upward trend despite periods of decline.

    * Are these harvest outcomes similar to other regions in Brazil?  

There are no similar outcomes in harvest in other regions of Brazil. There is some slight variation for certain periods. Regions like Espirito Santo, São Paulo, and Paraná have much lower and more consistent harvest outcomes. These other regions exhibit more stability with less drastic changes year over year. 

    * Which region produces the most coffee (by sub-category) in Brazil?  
From the chart, Minas Gerais produces significantly more coffee compared to the other regions in Brazil.

```python
temp_data = pd.read_csv("data/csv/merge_data")
temp_data["year"] = pd.to_datetime(temp_data["year"], format="%Y")
#temp_data["year"] = pd.to_datetime(temp_data["year"],format="%Y")

#plot weather changes throughout the years in Minas Gerais
plt.figure(figsize = (10,6))
sb.lineplot(data = temp_data,x = "year",y = "temperature_2m",marker="o")
plt.title("Temperature Changes Throughout the Years in Minas Gerais", fontsize = 16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Temperature", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
